```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('greendestination (1).csv')

# Calculate the attrition rate
total_employees = len(data)
employees_left = data[data['Attrition'] == 'Yes']
attrition_rate = (len(employees_left) / total_employees) * 100
print(f'Attrition Rate: {attrition_rate:.2f}%')

# Convert necessary columns to numeric
data['Age'] = pd.to_numeric(data['Age'], errors='coerce')
data['YearsAtCompany'] = pd.to_numeric(data['YearsAtCompany'], errors='coerce')
data['MonthlyIncome'] = pd.to_numeric(data['MonthlyIncome'], errors='coerce')

# Plot Age Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data=data, x='Age', hue='Attrition', multiple='stack', kde=True)
plt.title('Age Distribution by Attrition Status')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Print Age Statistics
print("Age Statistics by Attrition Status:")
print(data.groupby('Attrition')['Age'].describe())

# Plot Years at Company Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data=data, x='YearsAtCompany', hue='Attrition', multiple='stack', kde=True)
plt.title('Years at Company Distribution by Attrition Status')
plt.xlabel('Years at Company')
plt.ylabel('Count')
plt.show()

# Print Years at Company Statistics
print("Years at Company Statistics by Attrition Status:")
print(data.groupby('Attrition')['YearsAtCompany'].describe())

# Plot Monthly Income Distribution
plt.figure(figsize=(10, 5))
sns.histplot(data=data, x='MonthlyIncome', hue='Attrition', multiple='stack', kde=True)
plt.title('Monthly Income Distribution by Attrition Status')
plt.xlabel('Monthly Income')
plt.ylabel('Count')
plt.show()

# Print Monthly Income Statistics
print("Monthly Income Statistics by Attrition Status:")
print(data.groupby('Attrition')['MonthlyIncome'].describe())
```
